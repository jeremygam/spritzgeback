```python
### generates layer 1 ###

import svgwrite
from svgwrite import cm, mm
import csv

dwg = svgwrite.Drawing('spritz_hotswap_layer_1.svg', size=('313mm', '96mm'), viewBox=('0 0 313 96'))
hk = 0.1 #half kerf width
x_span = [-17, 296]
y_span = [-6.075, 89.925]
outer_width = (x_span[1] - x_span[0] + 2*hk)
outer_height = (y_span[1] - y_span[0] + 2*hk)

# outer cut outline
dwg.add(dwg.rect((x_span[0]-hk, y_span[0]-hk), (outer_width, outer_height), #slightly bigger for kerf
    stroke = 'black',
    fill = "none",
    rx = 2+hk,
    ry = 2+hk,
    stroke_width=0.2)
)

# mounting hole cuts
with open('Spritzgeback_hole_coordinates.csv', newline='') as csvfile:
    for line in csv.reader(csvfile):
        x = float(line[0])
        y = float(line[1])
        dwg.add(dwg.circle(center=(x,y),
            r=(3.96/2-hk-0.02), #subtract for kerf plus little extra
            stroke='black',
            fill='none',
            stroke_width=0.2)
        )

# cutout for USB
with open('Spritzgeback_USBcutout1_coordinates.csv', newline='') as csvfile:
    points = []    
    for line in csv.reader(csvfile):
        points.append((float(line[0]), float(line[1])))
dwg.add(dwg.polyline(points,
    stroke='black', fill='none',stroke_width=0.2))

    
dwg.save()
```

```python
# make fidget

dwg = svgwrite.Drawing('fidget_1p5.svg', profile='tiny')


import itertools
from itertools import product 

width = 1.5

rows = 6-1
cols = 5-1
row_centers = [(10+20*i) for i in range(0,rows+1)]
col_centers = [(10+20*i) for i in range(0,cols+1)]

centers = [(x,y) for x in row_centers for y in col_centers]

count = 0

for center in centers:
    x = center[0]
    y = center[1]
    points = [(x-6.8, y-6.8), (x+6.8, y-6.8), (x+6.8, y+6.8), (x-6.8, y+6.8), (x-6.8, y-6.8)]
    dwg.add(dwg.polyline(points,                     
        stroke='black', fill='none',stroke_width=0.2))

    points = [(x-10, y-10), (x+10, y-10), (x+10, y+10), (x-10, y+10), (x-10, y-10)]
    dwg.add(dwg.polyline(points,                     
        stroke='black', fill='none',stroke_width=0.2))

    if count%2 == 0:
        points = [(x-10-width, y-3), (x-10-width, y+3), (x+10+width, y+3), (x+10+width, y-3), (x-10-width, y-3)]
        dwg.add(dwg.polyline(points,                     
            stroke='black', fill='none',stroke_width=0.2))
    else: 
        points = [(x-3, y-10-width), (x+3, y-10-width), (x+3, y+10+width), (x-3, y+10+width), (x-3, y-10-width)]
        dwg.add(dwg.polyline(points,                     
            stroke='black', fill='none',stroke_width=0.2))
        
    count = count + 1
dwg.save()
```
